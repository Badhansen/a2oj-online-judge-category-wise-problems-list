```python
# The Code start here
"""
    Scrape the a2oj.com Online Judge Ladder using this code
    Change the url to you desire url
    and store it excel file 
"""
from bs4 import BeautifulSoup
import requests

url = "https://a2oj.com/category?ID=97"
catagory = "bitmasks problems"
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

# print(soup.prettify())

soup.prettify()
table_body = soup.find('tbody') # find all the table body data in the table_body list
table_rows = table_body.find_all('tr') # find all the table row data in table_rows list

i = int(1)
"""
    links dictionary will store the problem name as a key and problem oj link as value &
    problems dictionary will store the problem name as key and a list as value [source oj, difficulty level]
    
    why use seperate dictionary? why not one?
    -> Because the link is comming randomly so i need to map it any way
    
"""
links = {} #
problems = {}

for row in table_rows:
    cols = row.find_all('td') 
    t = row.find('a')
    # print("{0} {1}".format(i, cols))
    i = i + 1
    cols = [x.text.strip() for x in cols]
    #print("{0} {1}".format(t['href'], t.text))
    #print(t.text)
    links[t.text] = t['href']
    problems[cols[1]] = [cols[3], cols[6]]
# print(links)
data_base = []

#ptint(links)
#print(problems)
i = int(1)
for key, value in links.items():
    temp = []
    if key in problems.keys():
        temp.append(key)
        for item in problems[key]:
            temp.append(item)
        #print(problems[key])
        temp.append(value)
        data_base.append(temp)
        #print("{0} {1}".format(i, temp))
        i = i + 1
#print(data_base)


print("Total links : ", len(links))
print("Total problems : ", len(problems))

print("#########################################")

print("Total data Loss: ", len(links) - len(data_base))

data_base.insert(0, ["Problem Name", "Online Judge", "Difficulty Level", "Link"])

print(data_base[0])
print(data_base[2])


"""
    Storing the data into excel format
                as 
    ['Problem Name', 'Online Judge', 'Difficulty Level', 'Link']
"""
import xlwt


book = xlwt.Workbook(encoding="utf-8")
sheet1 = book.add_sheet(catagory)

i = 0

for problem in data_base:
    j = 0
    for item in problem:
        sheet1.write(i, j, item)
        j = j + 1
    i = i + 1
book.save(catagory + ".xls")
```
